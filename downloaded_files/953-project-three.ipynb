```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_bike = pd.read_csv('201805-citibike-tripdata.csv')
data_bike.head()
```

```python
data_bike.drop(data_bike.index[[1165053,1165074,1165078,1165097,1165106,1165108,1165128]])
```

```python
data_weather = pd.read_csv('NYC May 2018 Weather.csv')
data_weather.head()
```

```python
data_bike.dropna()
```

```python
from datetime import *
import time

hour = []
weekend = []
for start_time in data_bike['starttime']:
    t1 = datetime.strptime(start_time,'%Y-%m-%d %H:%M:%S')
    hour.append(t1.hour)
    if t1.weekday() in [5,6]:
        weekend.append(1)
    else:
        weekend.append(0)
```

```python
new_date_array = []

for start_time in data_bike['starttime']:
    t1 = datetime.strptime(start_time,'%Y-%m-%d %H:%M:%S')
    
    if ((len(str(t1.hour)) == 1) & (len(str(t1.day)) == 2) & (len(str(t1.month)) == 2)):
        new_date_array.append(str(t1.year)+'-'+str(t1.month)+'-'+str(t1.day)+'-0'+str(t1.hour))
        
    elif ((len(str(t1.day)) == 1) & (len(str(t1.month)) == 2) & (len(str(t1.hour)) == 2)):
        new_date_array.append(str(t1.year)+'-'+str(t1.month)+'-0'+str(t1.day)+'-'+str(t1.hour))
        
    elif ((len(str(t1.month)) == 1) & (len(str(t1.day)) == 2) & (len(str(t1.hour)) == 2)):
        new_date_array.append(str(t1.year)+'-0'+str(t1.month)+'-'+str(t1.day)+'-'+str(t1.hour))
        
    elif ((len(str(t1.month)) == 1) & (len(str(t1.day)) == 1) & (len(str(t1.hour)) == 2)):
        new_date_array.append(str(t1.year)+'-0'+str(t1.month)+'-0'+str(t1.day)+'-'+str(t1.hour))
        
    elif ((len(str(t1.month)) == 1) & (len(str(t1.day)) == 2) & (len(str(t1.hour)) == 1)):
        new_date_array.append(str(t1.year)+'-0'+str(t1.month)+'-'+str(t1.day)+'-0'+str(t1.hour))
        
    elif ((len(str(t1.month)) == 2) & (len(str(t1.day)) == 1) & (len(str(t1.hour)) == 1)):
        new_date_array.append(str(t1.year)+'-'+str(t1.month)+'-0'+str(t1.day)+'-0'+str(t1.hour))
    
    elif ((len(str(t1.month)) == 1) & (len(str(t1.day)) == 1) & (len(str(t1.hour)) == 1)):
        new_date_array.append(str(t1.year)+'-0'+str(t1.month)+'-0'+str(t1.day)+'-0'+str(t1.hour))
        
    elif ((len(str(t1.month)) == 2) & (len(str(t1.day)) == 2) & (len(str(t1.hour)) == 2)):
        new_date_array.append(str(t1.year)+'-'+str(t1.month)+'-'+str(t1.day)+'-'+str(t1.hour))
```

```python
clean_data_bike = pd.DataFrame({
    "date":data_bike['starttime'],\
    "duration":data_bike['tripduration'],\
    "weekend":weekend,\
    "hour":hour,\
    "start_lat":data_bike['start station latitude'],\
    "start_long":data_bike['start station longitude'],\
    "end_lat":data_bike['end station latitude'],\
    "end_long":data_bike['end station longitude'],\
    "gender":(data_bike['gender']-1),\
    "age":(2018 - data_bike['birth year']),\
    "new_date" : new_date_array
})
clean_data_bike.head()
```

```python
data_weather['new_date'] = data_weather['Date/Time'].str.extract('(../../.... ..)', expand=True)
data_weather.head()
```

```python
for i in range(0,len(data_weather['Date/Time'])):
    data_weather['new_date'][i] = data_weather['Date/Time'][i][6:10]+'-'+\
    data_weather['Date/Time'][i][0:2]+'-'+\
    data_weather['Date/Time'][i][3:5]+'-'+\
    data_weather['Date/Time'][i][11:13]
data_weather.head()
```

```python
cleaned_data_weather = pd.DataFrame({
    'Date-Time':data_weather['new_date'],
    'Air Temp F':data_weather['Air Temp (℉)'],
    'Wind Spd (mph)':data_weather['Wind Spd (mph)']
})
cleaned_data_weather.head()
```

```python
cleaned_data_bike = pd.DataFrame({
    'Date-Time':clean_data_bike['new_date'],
    'Age':clean_data_bike['age'],
    'Duration':clean_data_bike['duration'],
    'Weekend':clean_data_bike['weekend']
})
cleaned_data_bike_2 = cleaned_data_bike.dropna()
cleaned_data_bike_2.head()
```

```python
merged_df = pd.merge(cleaned_data_bike_2, cleaned_data_weather, on='Date-Time')
merged_df.head()
```

```python
# Added example for two output columns from a single input column
grouped_df = merged_df.groupby(['Date-Time']).agg({'Duration': np.mean, 'Air Temp F': np.mean,'Wind Spd (mph)': np.mean}) 
grouped_df.head()
```

```python
grouped_df.to_csv('Grouped_Data.csv')
```
