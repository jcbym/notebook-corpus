```python


SendInput = ctypes.windll.user32.SendInput
#https://gist.github.com/wgm89/1c2c0d3ff50f63df975a
# C struct redefinitions 
PUL = ctypes.POINTER(ctypes.c_ulong)
class KeyBdInput(ctypes.Structure):
    _fields_ = [("wVk", ctypes.c_ushort),
                ("wScan", ctypes.c_ushort),
                ("dwFlags", ctypes.c_ulong),
                ("time", ctypes.c_ulong),
                ("dwExtraInfo", PUL)]

class HardwareInput(ctypes.Structure):
    _fields_ = [("uMsg", ctypes.c_ulong),
                ("wParamL", ctypes.c_short),
                ("wParamH", ctypes.c_ushort)]

class MouseInput(ctypes.Structure):
    _fields_ = [("dx", ctypes.c_long),
                ("dy", ctypes.c_long),
                ("mouseData", ctypes.c_ulong),
                ("dwFlags", ctypes.c_ulong),
                ("time",ctypes.c_ulong),
                ("dwExtraInfo", PUL)]

class Input_I(ctypes.Union):
    _fields_ = [("ki", KeyBdInput),
                 ("mi", MouseInput),
                 ("hi", HardwareInput)]

class Input(ctypes.Structure):
    _fields_ = [("type", ctypes.c_ulong),
                ("ii", Input_I)]

# Actuals Functions

def PressKey(hexKeyCode):
    extra = ctypes.c_ulong(0)
    ii_ = Input_I()
    ii_.ki = KeyBdInput( 0, hexKeyCode, 0x0008, 0, ctypes.pointer(extra) )
    x = Input( ctypes.c_ulong(1), ii_ )
    ctypes.windll.user32.SendInput(1, ctypes.pointer(x), ctypes.sizeof(x))

def ReleaseKey(hexKeyCode):
    extra = ctypes.c_ulong(0)
    ii_ = Input_I()
    ii_.ki = KeyBdInput( 0, hexKeyCode, 0x0008 | 0x0002, 0, ctypes.pointer(extra) )
    x = Input( ctypes.c_ulong(1), ii_ )
    ctypes.windll.user32.SendInput(1, ctypes.pointer(x), ctypes.sizeof(x))
    
def left_click():
        extra = ctypes.c_ulong(0)
        ii_ = Input_I()
        ii_.mi = MouseInput(1, 1, 0, 0x0001, 0, ctypes.pointer(extra))
        x = Input(ctypes.c_ulong(0), ii_)
        ctypes.windll.user32.SendInput(1, ctypes.pointer(x), ctypes.sizeof(x))
        
        extra = ctypes.c_ulong(0)
        ii_ = Input_I()
        ii_.mi = MouseInput(0, 0, 0, 0x0002, 0, ctypes.pointer(extra))
        x = Input(ctypes.c_ulong(0), ii_)
        ctypes.windll.user32.SendInput(1, ctypes.pointer(x), ctypes.sizeof(x))
    
        extra = ctypes.c_ulong(0)
        ii_ = Input_I()
        ii_.mi = MouseInput(0, 0, 0, 0x0004, 0, ctypes.pointer(extra))
        x = Input(ctypes.c_ulong(0), ii_)
        ctypes.windll.user32.SendInput(1, ctypes.pointer(x), ctypes.sizeof(x))
        
def focusMouse():
    extra = ctypes.c_ulong(0)
    ii_ = Input_I()
    ii_.mi = MouseInput(2, 2, 0, 0x0001, 0, ctypes.pointer(extra))
    x = Input(ctypes.c_ulong(0), ii_)
    ctypes.windll.user32.SendInput(1, ctypes.pointer(x), ctypes.sizeof(x))
        
        
        
#while(True):
#    left_click()
#    time.sleep(1)
    

time.sleep(2)
LargeChestTopLeft = imagesearch("LargeChestHeight.png", 0.8)
#time.sleep(2)

LargeChestBottomRight = imagesearch("LargeChestBottomRight.png", 0.8)
#time.sleep(2)

InventoryTopLeft = imagesearch("LargeChestBottom.png", 0.8)
    
#time.sleep(2)

InventoryBottomRight = imagesearch("InventoryBottomRight.png", 0.8)
#time.sleep(2)

BlockLeft = imagesearch("BlockLeft.png", 0.8)
BlockRight = imagesearch("BlockRight.png", 0.8)
BlockTop = imagesearch("BlockTop.png", 0.8)
BlockBottom = imagesearch("BlockBottom.png", 0.8)

blockWidth = BlockRight[0] - BlockLeft[0]
blockHeight = BlockBottom[1] - BlockTop[1]





def EmptyInventory():
    PressKey(0x2A)
    for i in range(0, 3):
        for j in range(0, 9):
            pyautogui.moveTo(BlockTopLeft[0] + j*blockWidth + blockWidth/2, BlockTopLeft[1]+ i*blockHeight + blockHeight/2, 0.2)
            left_click()
            time.sleep(0.01)
            left_click()
    ReleaseKey(0x2A)
    
class Item:
    name = ""
    value = 0
    position = 0
    
    def __init__(self, name, value, position):
        self.name = name
        self.value = value
        self.position = position


        
PictureCompare = ("blankSpace.PNG", 
                  
                  "netherrack.png",
                  
                  "dirt.png", "cobblestone.PNG", "iron.png", "arrow.png", "stick.png", "gravel.png", "granite.png", "andenite.png", "coal.png", "diamond.png",
                  "bamboo.PNG", "torch.png", "powder.png", "jwstair.png", "sign.png", "sign2.png", "wfence.png", "stoneBricks.png", "lapiz.png",
                  "blazeRod.png", "redstone.png", "netherwart.png", "seeds.png", "leather.png", "flint.png",
                  "hat.png", "chest.png", "legs.png", "shoes.png", "dpick.png", "pickaxe.png", "daxe.png", "axe.png", "shovel.png", "bow.png", "sword.png", "crossbow.png",
                  
                  "lbucket.png", "wbucket.png",
                  
                  "eggs.png", "carrot.png",
                  "cbeef.png", "cmutton.png", "cpork.png", 
                  "pork.png",
                  
                  "end")

#1. Find what all objects in the inventory are
#2. Sort with mouse
def SortInventory():
    
    itemArray = []
    for count in range(0, 54):
        #Find area of screen to search images by
        im = TakeScreenShotOfItem(count)
        
        #Find What image it is
        for i in range(0, len(PictureCompare)):
            pic = PictureCompare[i]
            if(pic != "end"):
                location = imagesearcharea(pic, 0,0,0,0, 0.9, im)
                
                #Once successfully found add an Item to the itemArray
                if location[0] != -1:
                    
                    if i == 0:
                        itemArray.append(Item(pic, 9999999, count))
                    else:
                        itemArray.append(Item(pic, i, count))
                    print(pic, count)
                    break
            else:
                itemArray.append(Item(pic, i, count))
    
    
    
    
    sorted = False
    foundCount = 0
    
    while foundCount < 54:
        sorted = True
        minVal = itemArray[foundCount].value
        minLoc = foundCount
        
        for count in range(foundCount, 54):
            if(itemArray[count].value < minVal):
                minVal = itemArray[count].value
                minLoc = count
        
        if(minVal != itemArray[foundCount].value):
            firstPos = FindPositionInChest(foundCount)
            secondPos = FindPositionInChest(minLoc)
            MoveMouseToPosition(secondPos)
            #print("SHOULD RUN")
            time.sleep(0.01)
            left_click()
            time.sleep(0.01)
            MoveMouseToPosition(firstPos)
            time.sleep(0.01)
            left_click()
            time.sleep(0.01)
            MoveMouseToPosition(secondPos)
            time.sleep(0.01)
            left_click()
            time.sleep(0.01)

        #swap
        item = itemArray[minLoc]
        itemArray[minLoc] = itemArray[foundCount]
        itemArray[foundCount] = item
        foundCount = foundCount + 1
    print("Sorted.")

        
                
def MoveMouseToPosition(pos):
    pyautogui.moveTo(pos[0], pos[1], 0)
    focusMouse()
        
def FindPositionInChest(count):
    j = count % 9
    i = count // 9
    #print((i, j))
    return (LargeChestTopLeft[0] + j*blockWidth + blockWidth/2, LargeChestTopLeft[1]+ i*blockHeight + blockHeight/2)

def FindPositionTopLeftInChest(count):
    j = count % 9
    i = count // 9
    return (LargeChestTopLeft[0] + j*blockWidth, LargeChestTopLeft[1]+ i*blockHeight, LargeChestTopLeft[0] +((j+5)*blockWidth), LargeChestTopLeft[1]+((i+5)*blockHeight))

def TakeScreenShotOfItem(count):
    pos = FindPositionTopLeftInChest(count)
    #print(pos)
    return pyautogui.screenshot(region=(pos[0]-(blockWidth*0.1),pos[1]-(blockHeight*0.25), blockWidth*1.4, blockHeight*1.2))


#EmptyInventory()
SortInventory()
#TakeScreenShotOfItem(0).save("Test.png")
#TakeScreenShotOfItem(35).save("Test2.png")
#TakeScreenShotOfItem(30).save("Test3.png")
```
