```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
sy = pd.Series(iris.target, dtype="category")
sy = sy.cat.rename_categories(iris.target_names)
df['species'] = sy
```

```python
df.head()
```

```python
x=df["petal length (cm)"].values
x
```

```python

bins = np.linspace(-6, 6, 12 * 4 + 1)
ns, _ = np.histogram(x, bins=bins)

sample_mean = np.mean(x)
sample_median = np.median(x)
mode_index = np.argmax(ns)
sample_mode = 0.5 * (bins[mode_index] + bins[mode_index + 1])

sns.distplot(x, bins=bins)
plt.axvline(sample_mean, c='k', ls=":", label="mean")
plt.axvline(sample_median, c='k', ls="--", label="median")
plt.axvline(sample_mode, c='k', ls="-", label="mode")
plt.title("mean, median, mode")
plt.xlabel("x")
plt.legend()
plt.show()
```

### 표본평균과 확률변수의 관계

```python
#7.3.3
x1=np.random.normal(size=100)
x1_m=x1.mean()
```

```python
#표본평균 50개 구하기
ls_xm=[]
for i in range(50):
    x=np.random.normal(size=100)
    xmean=x.mean()
    ls_xm.append(xmean)
len(ls_xm)
```

```python
# 원래의 확률변수 분포
import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(x1)
plt.xlim(-5,5)
plt.show()
```

```python
#표본평균의 분포
sns.distplot(ls_xm)
plt.xlim(-2,2)
plt.show()
```

### 결합확률질량함수

```python
grades = ["A", "B", "C", "D", "E", "F"]
scores = pd.DataFrame(
    [[1, 2, 1, 0, 0, 0],
     [0, 2, 3, 1, 0, 0],
     [0, 4, 7, 4, 1, 0],
     [0, 1, 4, 5, 4, 0],
     [0, 0, 1, 3, 2, 0],
     [0, 0, 0, 1, 2, 1]], 
    columns=grades, index=grades)
scores.index.name = "Y"
scores.columns.name = "X"
scores
```

```python
#결합확률질량함수 :pmf
pmf= scores/scores.values.sum()
pmf
```

```python
# 주변확률질량함수: 두 확률변수 중 하나의 확률변수에 대해서마 확률분표 표시한것
marginal_x= pmf.sum(axis= 0)
marginal_x
```

```python
marginal_x= pmf.sum(axis= 1)
marginal_x
```

```python
# 조건부 질량함수: 확률변수 중 하나의 값이 특정값으로 고정된 경우 나머지 변수에 대한 확률질량함수
#y=A일때 결합확률질량함수
#p(x,y=A)
pmf.iloc[0,:]
```

```python
np.sum(pmf.iloc[0,:])
```

```python
#y=A일때 조건부확률질량함수와 확률의 합 (넓이가 1이 되도록 표준화)
# y=A일때 결합확률질량함수/ y=A일때의 x의 주변확률질량함수
#(x|y=A)
cond=pmf.iloc[0,:]/marginal_x[0]
cond
```
