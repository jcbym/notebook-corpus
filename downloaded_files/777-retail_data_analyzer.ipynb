Other Smaller therefore shows whatever is left, given the applied filters, after considering the shown top occurrences
<br><img src="https://www.dropbox.com/s/s2r7ucbisj8jl6q/rest_of_market.png?dl=1", width="85%", style="align:left"><br>

Weighted distribution info in faceted charts is available only as a "<i>Share of Total</i>" metric. Chose the "<i>Share</i>" metric if you want to see each facet's percentage of the sum of the metric for all facets.

```python
def make_directories():
    """ makes working folders if they do not already exist
    """
    !IF NOT EXIST tidy_data mkdir tidy_data
    !IF NOT EXIST CY_YA_charts mkdir CY_YA_charts
    return None

def save_downloaded_input_file(url,fileName):
    fileParams=get_file_params()
    directory=fileParams["directory"]
    u = urllib.request.urlopen(url)
    data = u.read()
    u.close()
    with open(directory+"/"+fileName, "wb") as f :
        f.write(data)
    return None

def drop_columns(df,toDrop):
    """ drops colums in list from dataframe
    """
    columns=list(df.columns.values)       
    for column in columns:
        if column in toDrop:
            df.drop(column, axis=1, inplace=True)
    return df

def delete_useless_columns(df): 
    """ drops colums that have only one distinct data element
    """    
    columns=list(df.columns.values)
    toDrop=[]
    for column in columns:
        if len(df[column].value_counts())<= 1:
            toDrop.append(column)
    df=drop_columns(df,toDrop)
    return df

def check_if_local_data_is_uptodate(dropboxMostRecentDataPath,localMostRecentDataPath):
    fileParams=get_file_params()
    mostRecentDataFileName=fileParams["mostRecentDataFileName"]
    mostRecentDataFileNameLocal=fileParams["mostRecentDataFileNameLocal"]
    dropboxMostRecentInfo=""
    localMostRecentInfo=""
    try:
        localMostRecentInfo=pd.read_csv(localMostRecentDataPath)
        localWeeks=localMostRecentInfo.loc[(localMostRecentInfo["Time Period"]=="Weeks")]
        localWeeks=localWeeks["Most Recent Data"].tolist()[0]
        try:
            dropboxMostRecentInfo=pd.read_csv(dropboxMostRecentDataPath)
            dropboxWeeks=dropboxMostRecentInfo.loc[(dropboxMostRecentInfo["Time Period"]=="Weeks")]
            dropboxWeeks=dropboxWeeks["Most Recent Data"].tolist()[0]
            if localWeeks==dropboxWeeks: 
                choice="local"
            else:
                choice="dropbox"
                save_downloaded_input_file(dropboxMostRecentDataPath,mostRecentDataFileNameLocal+".csv")
        except:
            choice="local"
    except:
        try:
            dropboxMostRecentInfo=pd.read_csv(dropboxMostRecentDataPath)
            save_downloaded_input_file(dropboxMostRecentDataPath,mostRecentDataFileNameLocal+".csv")
            choice="dropbox" 
        except:
            choice=None
    return choice,dropboxMostRecentInfo,localMostRecentInfo  

def open_input_files():
    """ opens input files. Tries to download from Dropbox. If error tries to download locally
    """    
    progress=0
    progress=make_progress_bar(progress)
    dfDict=collections.OrderedDict()
    fileParams=get_file_params()
    directory=fileParams["directory"]
    inFileNames=fileParams["inFileNames"]
    timePeriods=fileParams["timePeriods"]
    mostRecentDataFileName=fileParams["mostRecentDataFileName"]
    mostRecentDataFileNameLocal=fileParams["mostRecentDataFileNameLocal"]
    dropboxCode=fileParams["dropboxCode"]    
    dropboxLink=fileParams["dropboxLink"]    
    dropboxSuffix=fileParams["dropboxSuffix"] 
    datasets=fileParams["datasets"]
    printLog=stop_printing()
    mostRecentDataCode=dropboxCode["mostRecentData"]
    dropboxMostRecentDataPath=dropboxLink+mostRecentDataCode+"/"+mostRecentDataFileName+".csv"+dropboxSuffix
    localMostRecentDataPath=directory+"/"+mostRecentDataFileNameLocal+".csv"
    choice,dropboxMostRecentInfo,localMostRecentInfo=check_if_local_data_is_uptodate(dropboxMostRecentDataPath,localMostRecentDataPath)
    if choice == "dropbox":
        message='<p style="font-size:15px">Please be patient.......Loading data from the cloud......should not take too long....</p>'
        mostRecentInfo=dropboxMostRecentInfo
    elif choice=="local":
        message='<p style="font-size:15px">Please be patient.......Loading data from disk.....should be done in a moment.....</p>'
        mostRecentInfo=localMostRecentInfo
    else:
        mostRecentInfo=None
        message='<p style="font-size:15px">Cannot load local or remote data</p>'
        sys.exit(HTML(message))  
    display(HTML(message))  
    for timePeriod in timePeriods:
        dfDict[timePeriod]={}
        dropboxCode=fileParams["dropboxCode"][timePeriod]
        for dataset in datasets:
            dfDict[timePeriod][dataset]={}
            if choice == "dropbox":
                code=dropboxCode[dataset]
                fileName=inFileNames[dataset]+timePeriod+".csv"
                path=dropboxLink+code+"/"+fileName+dropboxSuffix
                try:
                    save_downloaded_input_file(path,fileName)
                except:
                    print("Unable to download from dropbox")
            else:
                path=directory+"/"+inFileNames[dataset]+timePeriod+".csv"
            try:
                df=pd.read_csv(path)
                df=delete_useless_columns(df)
                df.rename(columns={"TY_YA":"CY_YA","Sub_brand":"Sub Brand"}, inplace=True)
                df.loc[df["CY_YA"]=="TY","CY_YA"]="CY" 
                dfDict[timePeriod][dataset]=df
                progress.value += 1
                if printLog:
                    print("Opened input file",dataset,"..............")
            except:
                display(HTML('<p style="font-size:15px">could not get data. Link below not functioning</p>'))
                display(HTML(path))
                sys.exit() 
    display(HTML('<p style="font-size:15px">Data loaded successfully</p>'))
    display(HTML('<p style="font-size:15px">Most recent periods are:</p>'))    
    display(mostRecentInfo)       
    return dfDict,progress

def save_df_extract(df,filename,yMetric,facet):
    """ saves dataframe used to generate a given plot to csv
    """
    facet=facet.replace("by ","").strip()
    toDrop=[]
    columns=list(df.columns.values)
    keep=columns[0:3]
    if len(facet)>0:
        keep.append(facet)
    keep.append(yMetric)
    for column in columns:
        if column not in keep:
            toDrop.append(column)  
    df=drop_columns(df,toDrop)
    fileParams=get_file_params()
    printLog=stop_printing()
    directory=fileParams["directory"]
    outDirectory=fileParams["outDirectory"]
    path=outDirectory+"/"+filename+".csv"
    df.to_csv(path, index=False)
    return df
    
def group_shades(dfDict,orderDict):
    """ aggregates all hair colors in four groups
    """
    fileParams=get_file_params()
    datasetsWithShades=fileParams["datasetsWithShades"]
    generalParams=get_general_params()   
    subShades=generalParams["subShades"]
    printLog=stop_printing()
    for dataset in dfDict:        
        if dataset in datasetsWithShades:
            orderDict["Color"]=[]
            orderDict["Color Detail"]=[]
            df=dfDict[dataset]
            columns=list(df.columns.values)
            if "Color" in columns:
                df.insert(7,"Color Detail",df["Color"])
                for group in subShades:
                    if group!=None:
                        orderDict["Color"].append(group)
                    for subShade in subShades[group]:
                        df.loc[df["Color"]==subShade,"Color"]=group
                        orderDict["Color Detail"].append(subShade)
                dfDict[dataset]=df
                if printLog:
                    print("Aggregated shades in Brown, Black, Blonde and Other.........")    
    if "Color" in orderDict:
        orderDict["Color"].sort()
        orderDict["Color Detail"].sort()      
    return dfDict,orderDict

def change_product_type_classifications(dfDict):
    """ reclassifies type of some brands and sub brands
    """
    printLog=stop_printing()
    fileParams=get_file_params()
    datasetsWithTypes=fileParams["datasetsWithTypes"]
    generalParams=get_general_params()
    changeTypeLists=generalParams["changeTypeLists"] 
    for column in changeTypeLists:
        for dataset in dfDict:
            if dataset in datasetsWithTypes:
                df=dfDict[dataset]    
                columns=list(df.columns.values)
                if column in columns:
                    for changeTypeList in changeTypeLists[column]:
                        df.loc[df[column].isin(changeTypeLists[column][changeTypeList]),"Type"]=changeTypeList   
                dfDict[dataset]=df
                if printLog:
                    print("Changed incorrect product classifications.........") 
    return dfDict

def filter_data(dfDict,dataFilters):
    """ filters data based on user input
    """
    printLog=stop_printing()
    toDelete=[]
    for dataset in dfDict:
        filterKeys="filters"
        df=dfDict[dataset]  
        columns=list(df.columns.values)   
        for dataFilter in dataFilters:
            if dataFilters[dataFilter]:
                filterKeys=filterKeys+"_"+dataFilter+"-"+dataFilters[dataFilter]
                if dataFilter in columns:
                    df=df.loc[(df[dataFilter]==dataFilters[dataFilter])]
                    if printLog:
                        print("Filtered",dataset,"dataset by",dataFilters[dataFilter],dataFilter+"...............")    
                else:
                    df=df.loc[(df[columns[0]]=="999999999999999")]
                    toDelete.append(dataset)
                    if printLog:
                        print("Tagged",dataset,"dataset: not usable if filtering by",dataFilter+"...............") 
        dfDict[dataset]=df
    for item in toDelete:
        if item in dfDict: 
            del dfDict[item]
            if printLog:
                print("Deleted",item,"dataset from dictionary")  
    return dfDict,filterKeys


def get_top_per_facet(dfDict,orderDict,status,numberOfFacets,maxRank):
    """ gets top elements per facet as requested by user
    """
    generalParams=get_general_params() 
    facetList=generalParams["facetList"]
    facetParams=generalParams["facetParams"]
    fileParams=get_file_params()
    allFacetsDataset=fileParams["allFacetsDataset"]
    chosenTopDict={}
    excludedTopDict={}
    printLog=stop_printing()
    for facet in facetList:
        if status=="before":
            numberOfTop=20                
        elif status == "after":
            numberOfTop=numberOfFacets-1
        if facetParams[facet]["groupSmall"]:                  
            if facetParams[facet]["dataset"] in dfDict:
                df=dfDict[facetParams[facet]["dataset"]]
            else:
                df=dfDict[allFacetsDataset]
            groupByMetric=facetParams[facet]["groupByMetric"]
            dfGrouped=df.groupby(facet,as_index=False)[groupByMetric].sum()
            dfGrouped.sort_values(by=groupByMetric,ascending=False,inplace=True)
            chosenTopDict[facet]=dfGrouped[facet].head(32).tolist()
            if maxRank != None:
                excludedTopDict[facet]=chosenTopDict[facet][0:maxRank-1]
                chosenTopDict[facet]=chosenTopDict[facet][maxRank-1:numberOfTop+maxRank-1]              
            else:
                excludedTopDict[facet]=[] 
            orderDict[facet]=dfGrouped[facet].head(numberOfTop).tolist()     
            if printLog:
                print("Top ",facet+"s: ",chosenTopDict[facet],".........")
    return orderDict,chosenTopDict,excludedTopDict

def to_proper_case(dfDict):
    """ sets all names to capital case
    """
    generalParams=get_general_params() 
    facetList=generalParams["facetList"]
    printLog=stop_printing()
    for dataset in dfDict:
        df=dfDict[dataset]
        columns=list(df.columns.values)
        for facet in facetList:
            if facet in columns:
                df[facet]=df[facet].str.title()
        dfDict[dataset]=df
    if printLog:
        print("Set facets to Proper Case........")        
    return dfDict
        
def rename_smaller_to_rest_of_market(dfDict,orderDict,numberOfFacets,maxRank):
    """ renames lower revenue items to rest market
    """    
    generalParams=get_general_params() 
    facetParams=generalParams["facetParams"]
    orderDict,chosenTopDict,excludedTopDict=get_top_per_facet(dfDict,orderDict,"after",numberOfFacets,maxRank)
    printLog=stop_printing()
    for dataset in dfDict:
        df=dfDict[dataset]
        columns=list(df.columns.values)       
        for facet in chosenTopDict:
            if facet in columns:
                df.loc[(~df[facet].isin(chosenTopDict[facet])) & (~df[facet].isin(excludedTopDict[facet])),facet]=facetParams[facet]["RestName"]
                if chosenTopDict[facet] not in orderDict[facet]:
                    orderDict[facet]=chosenTopDict[facet]
            if printLog:
                print("Renamed smaller",facet+"s to",facetParams[facet]["RestName"],"in",dataset,"dataset........")                
            if facetParams[facet]["RestName"] not in orderDict[facet] and facetParams[facet]["makeRest"] and facetParams[facet]["RestName"] !=None:
                orderDict[facet].append(facetParams[facet]["RestName"])
        dfDict[dataset]=df
    return dfDict,orderDict

def find_index_columns(df,valueCols):
    columns=list(df.columns.values)
    indexCols=[]
    for column in columns:
        if column not in valueCols:
            indexCols.append(column)
    return indexCols        
    

def compress_rows(dfDict):
    """ compresses (aka groupsby) rows so we have totals for chart
    """    
    generalParams=get_general_params()
    indexCols={}
    valueCols=generalParams["compressRows"]["valueCols"] 
    aggDict=generalParams["compressRows"]["aggDict"] 
    for element in dfDict:  
        df=dfDict[element]          
        indexCols[element]=find_index_columns(df,valueCols[element])
        df=df.groupby(indexCols[element])[valueCols[element]].agg(aggDict[element]).reset_index()
        dfDict[element]=df       
    printLog=stop_printing()
    if printLog:
        print("Compressed dataset rows so price calculations work...........")
    return dfDict    

def prepare_faceted_dfs(dfDict,facetedBy):
    """ prepares dataframes with faceted info as requested by user
    """
    dfFacetDict={}
    generalParams=get_general_params()
    groupbyCols=generalParams["groupbyCols"] 
    valueCols=generalParams["compressRows"]["valueCols"] 
    aggDict=generalParams["compressRows"]["aggDict"] 
    for element in dfDict:
        df=dfDict[element]
        columns=list(dfDict[element].columns.values)       
        for facet in facetedBy:
            if facetedBy[facet] and facet in columns:
                indexCols=[]
                indexCols.append(facet)
                indexCols=indexCols+groupbyCols
                dfNew=df.groupby(indexCols)[valueCols[element]].agg(aggDict[element]).reset_index()
                dfFacetDict[facet+"_"+element]=dfNew
    for element in dfDict:
        df=dfDict[element]  
        df=df.groupby(groupbyCols)[valueCols[element]].agg(aggDict[element]).reset_index()
        dfDict[element]=df                
    for element in dfFacetDict:
        dfDict[element]=dfFacetDict[element]
    printLog=stop_printing()
    if printLog:
        print("Prepared dataframe for each facet...............")
    return dfDict

def get_prices(dfDict,country,currency):
    """ calculates prices
    """    
    generalParams=get_general_params()
    currencyDict=generalParams["currencyDict"] 
    promoDict=generalParams["promoDict"]
    multiplyDict=generalParams["multiplyDict"]
    for dataset in dfDict:
        df=dfDict[dataset] 
        columns=list(df.columns.values)
        df["Price"]=df[currencyDict[currency]]/df["Units"]*multiplyDict[country]        
        if "Promo_Units" in columns:
            df["Promo_Price"]=df[promoDict[currency]]/df["Promo_Units"]*multiplyDict[country]
            df["No_Promo_Price"]=(df[currencyDict[currency]]-df[promoDict[currency]])/(df["Units"]-df["Promo_Units"])*multiplyDict[country]
            df["Promo_Discount"]=(df["No_Promo_Price"]-df["Promo_Price"])/df["No_Promo_Price"]*100  
        if "Baseline_Units" in columns:
            df["Incremental_Value"]=df["Value_LC"]-df["Baseline_Value"]
            df["Incremental_Volume"]=df["Units"]-df["Baseline_Units"]
            df["Incremental_Value_%"]=df["Incremental_Value"]/df["Value_LC"]*100    
            df["Incremental_Volume_%"]=df["Incremental_Volume"]/df["Units"]*100 
            df["Trade_Deal_Efficiency"]=df["Incremental_Volume"]/df["Promo_Units"]*100 
            df["Trade_Deal_Effectiveness"]=df["Incremental_Value"]/df["Promo_LC"]*100 
            df["Promo_Elasticity"]=df["Incremental_Volume_%"]/df["Promo_Discount"] 
            df["Promo_Value_Elasticity"]=df["Incremental_Value_%"]/df["Promo_Discount"]       
            printLog=stop_printing()
            if printLog:
                print("Calculated prices for",dataset,"....................")
        df=dfDict[dataset] 
    return dfDict

def get_totals(df,columns,renameDict):
    """ gets total of filtered group
    """    
    generalParams=get_general_params()
    allValueCols=generalParams["compressRows"]["valueCols"]["shades_level"]
    groupbyCols=generalParams["groupbyCols"]
    valueCols,indexCols=[],[]            
    for element in columns:
        if element in allValueCols:
            valueCols.append(element)
        else:
            indexCols.append(element)                
    dfTotals=df.groupby(groupbyCols)[valueCols].sum()   
    dfTotals.rename(columns=renameDict, inplace=True)
    columnsdf=list(df.columns.values)
    columnsdfTotals=list(dfTotals.columns.values)
    for column in columnsdfTotals:
        if column in columnsdf:
            dfTotals.drop(column, axis=1, inplace=True)
    df=df.set_index(groupbyCols)
    df=df.join(dfTotals)
    df=df.reset_index()  
    return df

def get_market_totals(dfDict):
    """ gets total market values to calculate shares on total market
    """    
    generalParams=get_general_params()
    toDropCols=generalParams["toDropCols"]
    renameDict=generalParams["renameDict"]["beforeFiltered"]
    for dataset in dfDict:
        df=dfDict[dataset]
        columns=list(df.columns.values)
        df=get_totals(df,columns,renameDict)
        dfDict[dataset]=df
    return dfDict    

def get_shares(dfDict):
    """ calculates shares 
    """    
    generalParams=get_general_params()
    toDropCols=generalParams["toDropCols"]
    storesInUniverset=generalParams["storesInUniverse"]
    renameDict=generalParams["renameDict"]["afterFiltered"]
    for dataset in dfDict:
        df=dfDict[dataset]
        columns=list(df.columns.values)
        df=get_totals(df,columns,renameDict)
        df["Units_Market_Share"]=df["Units"]/df["Tot_Market_Units"]*100
        df["Value_LC_Share"]=df["Value_LC"]/df["Tot_LC"]*100
        df["Units_Share"]=df["Units"]/df["Tot_Units"]*100        
        if "Promo_Units" in columns:   
            df["Promo_USD_Market_Share"]=df["Promo_USD"]/df["Tot_Market_USD"]*100
            df["Promo_LC_Market_Share"]=df["Promo_LC"]/df["Tot_Market_LC"]*100
            df["Promo_Units_Market_Share"]=df["Promo_Units"]/df["Tot_Market_Units"]*100         
            df["Promo_USD_Share"]=df["Promo_USD"]/df["Tot_USD"]*100
            df["Promo_LC_Share"]=df["Promo_LC"]/df["Tot_LC"]*100            
            df["Promo_Units_Share"]=df["Promo_Units"]/df["Tot_Units"]*100           
            df["%_Promo_Value"]=df["Promo_USD"]/df["Value_USD"]*100
            df["%_Promo_Units"]=df["Promo_Units"]/df["Units"]*100
        if "Value_USD" in columns:     
            df["Value_USD_Market_Share"]=df["Value_USD"]/df["Tot_Market_USD"]*100
            df["Value_LC_Market_Share"]=df["Value_LC"]/df["Tot_Market_LC"]*100
            df["Value_USD_Share"]=df["Value_USD"]/df["Tot_USD"]*100            
        if "CWD" in columns:
            df["CWD_Share"]=df["CWD"]/df["Tot_CWD"]*100    
            df["CWD_Market_Share"]=df["CWD"]/df["Tot_Market_CWD"]*100                                     
            df["Rate_Of_Sales_Ratio"]=df["Value_LC_Market_Share"]/df["CWD_Market_Share"]             
        if "TDP" in columns:
            df["TDP_Share"]=df["TDP"]/df["Tot_TDP"]*100    
            df["TDP_Market_Share"]=df["TDP"]/df["Tot_Market_TDP"]*100
            df["Average_Items_Carried"]=df["TDP"]/df["CWD"]             
        df=drop_columns(df,toDropCols)
        dfDict[dataset]=df
        printLog=stop_printing()
        if printLog:
            print("Calculated shares for",dataset,"....................")
    return dfDict

def prepare_data(dataFilters,dfDict,orderDict,numberOfFacets,maxRank):  
    """ filters data based on user choices, renames smaller items to rest of market, and groups by
    """
    dfDict,filterKeys=filter_data(dfDict,dataFilters)
    dfDict,orderDict=rename_smaller_to_rest_of_market(dfDict,orderDict,numberOfFacets,maxRank)
    dfDict=compress_rows(dfDict)
    return dfDict,filterKeys,orderDict

def transform_data(dfDict,country,currency,facetedBy):
    """ prepares facets, calculates shares, gets prices
    """    
    dfDict=prepare_faceted_dfs(dfDict,facetedBy)
    dfDict=get_shares(dfDict)
    dfDict=get_prices(dfDict,country,currency)  
    return dfDict

def prepare_and_transform_data(newRunParams,dfDictOriginal,orderDictOriginal,country,currency): 
    """ does file preparation required for plotting a given chart
    """
    dfDict=copy.deepcopy(dfDictOriginal)
    orderDict=copy.deepcopy(orderDictOriginal)
    numberOfFacets,maxRank=newRunParams["numberOfFacets"],newRunParams["maxRank"]
    dataFilters,facetsToPlot=newRunParams["dataFilters"],newRunParams["facetsToPlot"] 
    dfDict,filterKeys,orderDict=prepare_data(dataFilters,dfDict,orderDict,numberOfFacets,maxRank)         
    dfDict=transform_data(dfDict,country,currency,facetsToPlot)
    return dfDict,filterKeys,orderDict,maxRank

def open_and_prepare_files():
    """ opens files and does initial cleaning (only done at initial file load)
    """
    #try:
    make_directories() 
    dfDict,progress=open_input_files()
    """
    except: 
        sys.exit("Unable to get file neither locally nor from cloud. Check internet connection and data folder")            
    """
    fileParams=get_file_params()
    
    timePeriods=fileParams["timePeriods"]
    orderDict={}
    for timePeriod in timePeriods:
        orderDict[timePeriod]={} 
        dfDict[timePeriod]=to_proper_case(dfDict[timePeriod])    
        dfDict[timePeriod],orderDict[timePeriod]=group_shades(dfDict[timePeriod],orderDict[timePeriod])    
        dfDict[timePeriod]=change_product_type_classifications(dfDict[timePeriod])
        dfDict[timePeriod]=get_market_totals(dfDict[timePeriod]) 
        orderDict[timePeriod],chosenTopDict,excludedTopDict=get_top_per_facet(dfDict[timePeriod],orderDict[timePeriod],"before",None,None)
        progress.value += 1
    progress.close()
    return dfDict,orderDict

def set_seaborn_settings(plotType):
    """ loads standard seaborn parameters
    """
    mpl.pyplot.close("all");
    seabornParams={"Standard":
                       {"lineWidth":1,"markerSize":1,"context":"notebook","fontScale":1},
                   "Faceted":
                       {"lineWidth":0.5,"markerSize":0.5,"context":"paper","fontScale":1}}
    paper_rc = {'lines.linewidth':seabornParams[plotType]["lineWidth"], 
                'lines.markersize':seabornParams[plotType]["markerSize"],}           
    sns.set_style("whitegrid",{'grid.linestyle': 'dotted',});
    sns.set_context(seabornParams[plotType]["context"],
                    font_scale=seabornParams[plotType]["fontScale"],rc = paper_rc);
    sns.despine(); 
    
def calculate_index_metric(df,dataset):
    """ calculates index (base 100) metrics
    """    
    generalParams=get_general_params()
    valueCols=generalParams["compressRows"]["valueCols"]["All"]
    potentialIndexCols=find_index_columns(df,valueCols)
    periods=["CY","YA"]
    toDrop=["CY_YA"]
    frames=[]
    count=0 
    for period in periods:
        indexCols=[]
        valueCols=[]
        frames.append(df.loc[(df["CY_YA"]==period)].copy())
        frames[count]=drop_columns(frames[count],toDrop)
        columns=list(frames[count].columns.values) 
        for column in columns:
            if column in potentialIndexCols:
                indexCols.append(column)
            else:
                valueCols.append(column)
        newValueCols=[]
        for column in valueCols:
            column=column+"_"+periods[count]
            newValueCols.append(column)
            toDrop.append(column)        
        frames[count].set_index(indexCols,inplace=True)  
        frames[count].columns = newValueCols
        count=count+1
    dfIndex=frames[0].join(frames[1])
    for element in valueCols:
        dfIndex[element]=(frames[0][element+"_"+periods[0]]/frames[1][element+"_"+periods[1]])*100
    dfIndex=drop_columns(dfIndex,toDrop)
    dfIndex.reset_index(inplace=True)
    return dfIndex



def plot_charts(dfPlot,yMetric,check,chartDict,tables,currency,filterKeys,facet,
                timePeriod,dataset,metricType,maxRank,numberOfFacets):
    """ does the actual plotting
    """
    metricType=get_choice(metricType)
    df=dfPlot.copy()
    generalParams,fileParams=get_general_params(),get_file_params()
    outDirectory,directory=fileParams["outDirectory"],fileParams["directory"]   
    periodDict=chartDict["periodDiv"]
    titleDict=chartDict["title"]
    if yMetric in ["Price"]:
        suffix=" in "+currency
    else:
        suffix=""
    columns=list(df.columns.values)
    for period in periodDict[yMetric] :
        if period in chartDict["order"][timePeriod] and "Period" in columns:
            df.loc[df["Period"] == period, yMetric] = df[yMetric]/periodDict[yMetric][period]          
    if metricType == "Index":
        df=calculate_index_metric(df,dataset)
        chartDict["hue"]=None
        chartDict["palette"]=None
        chartDict["color"]="#f4a582"
        yLim=[None,None]
        metricType = "- Index"
        yLabels="Index vs YA"
    else:
        yLim=chartDict["yLim"][yMetric]
        yLabels=chartDict["yLabels"][yMetric]+suffix
        metricType=""
    plt.figure(figsize=(12,9))  
    g=sns.factorplot(x=chartDict["xMetric"],y=yMetric, data=df, 
                     col=facet,col_order=chartDict["colOrder"],
                     palette=chartDict["palette"], hue=chartDict["hue"],color=chartDict["color"],
                     order=chartDict["order"][timePeriod],col_wrap=chartDict["colWrap"],
                     size=chartDict["size"], aspect=chartDict["aspect"],);
    g.fig.subplots_adjust(top=chartDict["topAdjust"]);
    g.despine(left=True);    
    g.set_axis_labels(chartDict["xLabel"],yLabels);    
    g.set(ylim=yLim);
    noFilter=""
    if filterKeys == "filters":
        noFilter = "_None"
    filename=titleDict[yMetric]+"_facet_"+str(facet)+"_"+filterKeys+noFilter+"_period_"+timePeriod+"_facets_"+str(numberOfFacets)+"_maxRank_"+str(maxRank)+metricType
    filterKeys=filterKeys.replace('filters_', '').replace('filters', '').replace('_',', ').replace('-',': ')
    if facet:
        facet= " by "+facet
    else:
        facet = ""
    title=chartDict["title"][yMetric]+str(facet)+". "+filterKeys+" "+metricType+check  
    g.fig.suptitle(title,size=chartDict["titleSize"],alpha=.7);  
    filename=filename.replace(' ', '_').replace('-','')
    path=outDirectory+"/"+filename       
    try:  
        plt.savefig(path+'.png', transparent=True)  
        datatable=save_df_extract(df,filename,yMetric,facet)
        tables[title]=datatable
    except:
        print("Error in saving chart image or csv extract file")
        print("Check if 'CY_YA_charts' directory exists or if csv file is open")
        pass
    plt.show();
    return tables

def plot_standard_charts(dfDict,category,country,orderDict,filterKeys,check,
                      tables,currency,newRunParams,maxRank):
    """ manages parameters to plot standard charts
    """    
    facet=None
    metricsToPlot,timePeriod,metricType=newRunParams["metricsToPlot"],newRunParams["timePeriod"],newRunParams["metricType"]
    timePeriod=get_choice(timePeriod)
    fileParams,chartParams=get_file_params(),get_chart_params(orderDict,timePeriod) 
    datasets,datasetContent=fileParams["datasets"],fileParams["datasetContent"]
    set_seaborn_settings("Standard")
    for metric in metricsToPlot:
        chartDict=chartParams[metric]
        if metricsToPlot[metric]:
            count=0
            for dataset in datasets:
                if dataset in dfDict and metric in datasetContent[dataset] and count==0:
                    count=count+1
                    dfPlot=dfDict[dataset]
                    numberOffacets=1
                    chartDict=set_chart_dict_to_standard(chartDict)
                    if metric=="Price Ladder":
                        dfPlot,metricCols=prepare_df_for_price_ladder(dfPlot)
                        chartDict["yMetrics"]=metricCols
                    if len(dfPlot)>0 and len(chartDict["yMetrics"])>0:
                        for yMetric in chartDict["yMetrics"]:
                            tables=plot_charts(dfPlot,yMetric,check,chartDict,tables,
                                                  currency,filterKeys,facet,timePeriod,dataset,metricType,maxRank,numberOffacets)
                    else:
                         print("Nothing to plot - Check filters/radio buttons or try restarting the notebook..........")
    return tables      

def plot_faceted_charts(dfDict,category,country,orderDict,filterKeys,check,tables,currency,newRunParams,maxRank):
    """ manages parameters to plot faceted charts
    """
    metricsToPlot,facetsToPlot,timePeriod,metricType=newRunParams["metricsToPlot"],newRunParams["facetsToPlot"],newRunParams["timePeriod"],newRunParams["metricType"]
    timePeriod=get_choice(timePeriod)
    fileParams,chartParams=get_file_params(),get_chart_params(orderDict,timePeriod) 
    datasets,datasetContent=fileParams["datasets"],fileParams["datasetContent"]
    set_seaborn_settings("Faceted")
    facetDatasets=[]
    facetDict={}
    for dataset in datasets:
        for facet in facetsToPlot:
            facetDatasets.append(facet+"_"+dataset)
            facetDict[facet+"_"+dataset]=dataset   
    for metric in metricsToPlot:
        chartDict=chartParams[metric]
        if metricsToPlot[metric]:
            for facet in facetsToPlot:
                count=0
                if facetsToPlot[facet]:
                    for dataset in facetDatasets:
                        if dataset in dfDict and metric in datasetContent[facetDict[dataset]] and facet in dataset and count==0:
                            count=count+1
                            dfPlot=dfDict[dataset]
                            numberOfFacets=len(orderDict[facet])                                                         
                            chartDict=set_chart_dict_to_facet(chartDict,orderDict,facet,filterKeys,numberOfFacets)
                            if metric=="Price Ladder":
                                dfPlot,metricCols=prepare_df_for_price_ladder(dfPlot)
                                chartDict["yMetrics"]=metricCols
                            if len(dfPlot)>0 and len(chartDict["yMetrics"])>0:                                
                                for yMetric in chartDict["yMetrics"]:
                                    tables=plot_charts(dfPlot,yMetric,check,chartDict,tables,currency,
                                                       filterKeys,facet,timePeriod,dataset,metricType,maxRank,numberOfFacets)                             
                            else:
                                print("Nothing to plot - Check filters/radio buttons or try restarting the notebook..........")
    return tables

def prepare_df_for_price_ladder(df):
    """ prepares data for price ladder chart
    """
    generalParams=get_general_params()
    allValueCols=generalParams["compressRows"]["valueCols"]["All"]
    priceValueCols=generalParams["compressRows"]["valueCols"]["Price Ladder"]
    renameDict=generalParams["renameDict"]["Price Ladder"]
    columns=list(df.columns.values)
    idVars=[]
    toDrop=[]
    dfLadder=df.copy()
    dfLadder.rename(columns=renameDict, inplace=True)
    for element in allValueCols:
        if element in columns and element not in priceValueCols:
            toDrop.append(element)
    dfLadder=drop_columns(dfLadder,toDrop)
    columns=list(dfLadder.columns.values) 
    for element in columns:
        if element not in priceValueCols:
             idVars.append(element)
    dfLadder,metricCols=melt_and_unmelt_dataframe(dfLadder,idVars,priceValueCols) 
    return dfLadder,metricCols

def melt_and_unmelt_dataframe(df,idVars,metrics):
    """ melts and unmelt price ladder data
    """
    varName="Metric"
    valueName="Value"
    periodName='Period'
    metricCols=[]
    df= pd.melt(df,id_vars=idVars, var_name=varName,
        value_vars=metrics,value_name="Value")
    idVars.remove(periodName)
    idVars.append(varName)    
    df=df.pivot_table(index=idVars,columns=[periodName],values=[varName])
    df.columns = df.columns.droplevel(0).rename(None)
    df=df.reset_index()
    columns=list(df.columns.values)
    for element in columns:
        if element not in idVars:
            metricCols.append(element)
    return df,metricCols


def start_plotting(dfDict,check,currency,tables,filterKeys,category,country,orderDict,newRunParams,maxRank): 
    """ manages different parameters for choice of standard and faceted charts
    """
    printLog=stop_printing()
    plotType=newRunParams["plotType"]
    if plotType["Standard"]:        
        if printLog:
            print("Started plotting total charts.............")
        tables=plot_standard_charts(dfDict,category,country,orderDict,filterKeys,check,
                          tables,currency,newRunParams,maxRank)
        if printLog:
            print("Finished plotting total charts.........")
    if plotType["Faceted"]:
        if printLog:
            print("Started plotting faceted charts.............")
        tables=plot_faceted_charts(dfDict,category,country,orderDict,filterKeys,check,
                            tables,currency,newRunParams,maxRank)
        if printLog:
            print("Finished plotting faceted charts.............")
    return None

def make_radio_buttons():
    """ creates radio buttons for choice of chart setup
    """
    radioGroups=[]
    checkDict={}
    generalParams=get_general_params()
    groups=generalParams["radioButtonParams"]["groups"]
    elementOrder=generalParams["radioButtonParams"]["elementOrder"]
    titleDict=generalParams["radioButtonParams"]["titleDict"] 
    for group in groups:
        options=[]
        choices=elementOrder[group]
        for choice in choices:
            options.append(choice)
        radio=widgets.RadioButtons(
        options=options,
        description=titleDict[group],
        disabled=False,
        )
        checkDict[group]=radio
        radioGroups.append(radio)
        #display(checkDict[group])
    return checkDict,radioGroups

def make_integer_sliders():
    """ creates the widgets that return the user's choice of number of 
    facets and of max rank to show
    """ 
    sliderGroups=[]
    sliderDict={}
    generalParams=get_general_params()
    groups=generalParams["intSliderParams"]["groups"]
    value=generalParams["intSliderParams"]["value"]
    minValue=generalParams["intSliderParams"]["minValue"]
    maxValue=generalParams["intSliderParams"]["maxValue"]    
    titleDict=generalParams["intSliderParams"]["titleDict"] 
    for group in groups:
        slider=widgets.IntSlider(
            value=value[group],
            min=minValue[group],
            max=maxValue[group],
            step=1,
            description=titleDict[group],
            disabled=False,
            continuous_update=False,
            orientation='horizontal',
            readout=True,
            readout_format='',
            slider_color='white'
        )
        sliderGroups.append(slider)
        sliderDict[group]=slider
        #display(sliderDict[group])
    return sliderDict,sliderGroups

def make_progress_bar(progress):
    """ creates the progress bar that shows status of file loading
    """
    progress=widgets.IntProgress(
    value=progress,
    min=0,
    max=12,
    step=1,
    description='Loading:',
    bar_style='info',
    orientation='horizontal'
    )
    display(progress)
    return progress

def batch_plot_widget():
    batchPlot=widgets.Checkbox(
        value=False,
        description='Enable batch plot',
        disabled=False
        )
    display(batchPlot)
    return batchPlot

def make_text_input(orderDict):
    """ creates the text widget for filtering choice
    """    
    filterDict={}
    filterGroups=[]
    generalParams=get_general_params()
    groups=generalParams["textInputParams"]["groups"]
    placeholders=generalParams["textInputParams"]["placeholders"]
    descriptions=generalParams["textInputParams"]["descriptions"]
    for group in groups:
        if None not in orderDict[group]:
            orderDict[group].append(None)
        textInput=widgets.Dropdown(
            options=orderDict[group],
            value=None,
            description=descriptions[group],
            disabled=False
            )
        filterGroups.append(textInput)
        filterDict[group]=textInput
        #display(filterDict[group])
    return filterDict,filterGroups

def get_choice(dictionary):
    """ pulls out the choice for a given widget from the dictionary
    """
    for element in dictionary:
        if dictionary[element]:
            choice=element
    return choice

def load_widget_values(checkDict,filterDict,sliderDict):
    """ pulls values from dictionaries returned from widgets and 
        and updates runParams dictionary
    """
    generalParams=get_general_params()
    runParams=generalParams["runParams"]
    for element in filterDict:
        if filterDict[element].value=="":
            runParams["dataFilters"][element]=None
        else:
            runParams["dataFilters"][element]=filterDict[element].value
    for element in checkDict:
        runParams[element][checkDict[element].value]=True
    for element in sliderDict:
        runParams[element]=sliderDict[element].value
    return runParams


def open_batch_plotting_file():
    """
    read excel file with plots to batch plot return dataframe
    """
    fileParams=get_file_params()
    batchPlotFileName=fileParams["batchPlotFileName"]
    batchPlotSheetName=fileParams["batchPlotSheetName"]
    df=pd.read_excel(batchPlotFileName+".xlsx",sheetname=batchPlotSheetName,
            skiprows=1,encoding='utf-8',)
    return df
        
def make_batch_plotting_dictionary(df): 
    """
    clean and transform batch plot dataframe in dictionary
    """    
    generalParams=get_general_params()
    required=generalParams["radioButtonParams"]["groups"][0:-1]
    intSliderParams=generalParams["intSliderParams"]
    widgetGroups=generalParams["widgetGroups"]
    widgetDfSlices=generalParams["widgetDfSlices"]
    chartsToPlotDict={}   
    df=df.loc[~((df["plotType"]=="Faceted") & (df["facetsToPlot"].isnull()))]
    df=df.dropna(subset=required)
    resultDict={}
    for element in widgetGroups:
        dfSlice=df.ix[:,widgetDfSlices[element][0]:widgetDfSlices[element][1]].copy()
        resultDict[element]=dfSlice.to_dict('index')
        for plot in resultDict[element]:
            for metric in resultDict[element][plot]:
                if pd.isnull(resultDict[element][plot][metric]):
                    resultDict[element][plot][metric]=None
    return resultDict

def load_batch_values(resultDict):
    """
    transforms resultDict in format for plotting
    """
    generalParams=get_general_params()
    runParams=generalParams["runParams"]
    widgetGroups=generalParams["widgetGroups"]
    batchParams={}
    for plotNumber in resultDict[widgetGroups[0]]: 
        batchParams[plotNumber]=copy.deepcopy(runParams)
        for element in resultDict[widgetGroups[0]][plotNumber]:
            choice=resultDict[widgetGroups[0]][plotNumber][element]            
            batchParams[plotNumber][element][choice]=True
        for element in resultDict[widgetGroups[1]][plotNumber]:
            if pd.notnull(resultDict[widgetGroups[1]][plotNumber][element]):
                batchParams[plotNumber][element]=int(resultDict[widgetGroups[1]][plotNumber][element])                               
        for element in resultDict[widgetGroups[2]][plotNumber]: 
            choice=resultDict[widgetGroups[2]][plotNumber][element]
            if resultDict[widgetGroups[2]][plotNumber][element]!= None:
                batchParams[plotNumber]["dataFilters"][element]=choice
    return batchParams       

def run_all(ev):
    display(Javascript('IPython.notebook.execute_cells_below()'))
    
def run_cell(ev):
    display(Javascript('IPython.notebook.execute_cell()'))
    

def batch_plot(dfDict,orderDict):
    df=open_batch_plotting_file()
    resultDict=make_batch_plotting_dictionary(df)
    batchParams=load_batch_values(resultDict)
    for element in batchParams:
        main(dfDict,orderDict,batchParams[element])

def stop_printing():
    """ turns on and off printing of warning messages
    """
    printLog=False
    return printLog

        
def main(dfDict,orderDict,newRunParams):
    timePeriods=newRunParams["timePeriod"]
    timePeriod=get_choice(timePeriods)
    check=""
    currency="LC"
    category="coloring"
    country="UK"
    tables={}
    printLog=stop_printing()   
    dfDictPeriod,filterKeys,orderDict[timePeriod],maxRank=prepare_and_transform_data(newRunParams,dfDict[timePeriod],
                                                                 orderDict[timePeriod],country,currency)    
    start_plotting(dfDictPeriod,check,currency,tables,filterKeys,category,country,
                   orderDict[timePeriod],newRunParams,maxRank)   
    if printLog:
        print("Saved files")
    return None


dfDict,orderDict=open_and_prepare_files()
```
