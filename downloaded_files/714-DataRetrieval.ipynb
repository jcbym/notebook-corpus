```python
import numpy as np
import xarray as xr
from siphon.catalog import TDSCatalog
import matplotlib.pyplot as plt
```

```python
url = 'http://smode.whoi.edu:8080/thredds/dodsC/insitu/drifters/'

base_url = 'http://smode.whoi.edu:8080/thredds/dodsC/insitu/drifters/'
cat = TDSCatalog('http://smode.whoi.edu:8080/thredds/catalog/insitu/drifters/catalog.html')
```

```python
# First, let's take a look single data file
drifter_data = xr.open_dataset(base_url+'drifter_0-4436220.nc')
```

```python
drifter_data
```

```python
# Get url for all data from all gliders
filelist=[file for file in cat.datasets]
datapath = [base_url+file for file in filelist]
```

```python
datapath
```

```python
# Try plotting trajectories for all drifters

fig, ax = plt.subplots(figsize=(10,10))

kwargs = {
    'color': 'k',
    'marker': '+',
    'ax': ax
}

for path in datapath:
    data = xr.open_dataset(path)
    data.plot.scatter(x='longitude',
                      y='latitude',
                      **kwargs)

ax.set_aspect(1/np.sin(37*np.pi/180))
```
