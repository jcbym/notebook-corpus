```python
# Import all the necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
# Read the Breast cancer data from csv
data = pd.read_csv('data.csv')
```

```python
data.head(10)
```

```python
# Feature Selection
x = pd.DataFrame(data.iloc[:,2:-5])  
x
```

```python
# Target Selection
# Diagnosis column would be our target
y = pd.DataFrame(data.iloc[:,1:2])
y

# Label Encoding
# This will conver the values M-1 and B-0
label = LabelEncoder()
print(y.shape)
y = label.fit_transform(y)

# Make a dataframe for target
y = pd.DataFrame(y, columns=['diagnosis'])
```

```python
# Splitting our dataset for trainnig and testing
x_train, x_test, y_train, y_test = train_test_split(x.values, y.values, test_size=0.20, random_state=24)

print('Shape of x_train: ', x_train.shape)
print('Shape of x_test: ', x_test.shape)
print('Shape of y_train: ', y_train.shape)
print('Shape of y_test: ', y_test.shape)
```

```python
# Creating the Logistic Regression model
logReg = LogisticRegression()

# Training the Logistic Regression model
logReg.fit(x_train, y_train.ravel())
```

```python
logReg.intercept_
```

```python
# Check the score of the model
logReg.score(x_test,y_test)
```

```python
# Evaluating the Logistic Regression Model
from sklearn.metrics import accuracy_score

logi_pred = logReg.predict(x_test)
acc_logisreg = accuracy_score(y_test, logi_pred)
print('Accuracy of the model is ', acc_logisreg)
```

# Second model

```python
# Import libraries for our second model - Random Forest
from sklearn.ensemble import RandomForestClassifier
```

```python
# Creating a random forest model
rfc_model = RandomForestClassifier(n_estimators=50, max_depth=2, random_state=0)

# Training the model
rfc_model.fit(x_train, y_train.ravel())
```

```python
# Check the score of the model
rfc_model.score(x_test, y_test)
```

```python
# Evaluating the Random Forest model accuracy

rfc_pred = rfc_model.predict(x_test)
accuracy_rfc = accuracy_score(y_test, rfc_pred)
print('Accuracy of the Random forest model is ', acc_logisreg)
```
