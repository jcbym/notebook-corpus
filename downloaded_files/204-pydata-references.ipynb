```python
def custom_style(alpha=.3):
    ax = plt.gca()
    
    for spine in ["bottom", "left"]:
        s = ax.spines[spine]
        s.set_position('zero')
        s.set_color('black')
        s.set_alpha(alpha)

    for spine in ["top", "right"]:
        s = ax.spines[spine]
        s.set_color('none')

    #ax.xaxis.set_ticks_position('bottom')
    #ax.yaxis.set_ticks_position('left')
    
    sns.set_style('whitegrid')
```

```python
X = np.linspace(-3,3)
plt.plot(X, np.exp(X));
custom_style()
```

```python
X = np.linspace(-3,3)
plt.plot(X, np.exp(-X));
custom_style()
```

```python
X = np.linspace(0.01, 5, num=200)
plt.plot(X, np.log(X));
custom_style()
```

```python
X = np.linspace(-5, -0.01, num=500)
plt.plot(X, np.log(-X));
custom_style()
```

```python
X = np.linspace(-5, 0, num=500)
l, *_ = plt.plot(X, 1/X);
X = np.linspace(0, 5, num=500)
plt.plot(X, 1/X, color=l.get_color());
custom_style()
plt.ylim(-10,10)
```
