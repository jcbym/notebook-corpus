```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.utils import normalize
import cv2 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## Load and Pre-process data

```python
# Bring in X_train and X_test data previously saved as numpy arrays
X_train = np.load('./Data/xtraindata.npy')
X_test = np.load('./Data/xtestdata.npy')
```

```python
X_train.shape
```

```python
# sample image
plt.imshow(X_train[0])
```

```python
trainlabels = pd.read_csv('./Labels/ISBI2016_ISIC_Part3_Training_GroundTruth.csv', header=None)
testlabels = pd.read_csv('./Labels/ISBI2016_ISIC_Part3_Test_GroundTruth.csv', header=None)
```

```python
trainlabels
```

```python
# Per ISIC (https://challenge.isic-archive.com/landing/2016/39), benign = 0, malignant = 1.
def changelabel(label):
    if label == 'benign':
        return 0
    elif label == 'malignant':
        return 1

# change the trainlabels to match testlables. Convert to numpy array.
y_train = trainlabels[1].map(changelabel).to_numpy().astype(np.float32)
y_train.shape
```

```python
# convert testlabels to numpy array
y_test = testlabels[1].to_numpy().astype(np.float32)
y_test.shape
```

```python
# normalize np array values between 0 and 1
X_train = normalize(X_train, axis=1)
X_test = normalize(X_test, axis=1)
```

```python
# sample new value
X_train[3].max()
```

## Build CNN model

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, Activation
from tensorflow.keras.optimizers import Adam
```

```python
# define the model
im_rows = 360
im_cols = 480
im_shape = (im_rows, im_cols, 1)

X_train = X_train.reshape(X_train.shape[0], *im_shape)
X_test = X_test.reshape(X_test.shape[0], *im_shape)

# confirm shape
print('x_train shape: {}'.format(X_train.shape))
print('x_test shape: {}'.format(X_test.shape))
```

```python
cnn_model = Sequential([
    Conv2D(filters=32, kernel_size=3, activation='relu', padding='same', kernel_initializer='glorot_uniform', input_shape=im_shape,),
    MaxPooling2D(pool_size=2),
    Dropout(0.2),

    Conv2D(filters=64, kernel_size=3, activation='relu'),
    MaxPooling2D(pool_size=2),
    Dropout(0.3),
    
    Flatten(),
    Dense(32, activation='relu'),
    Dense(2, activation='softmax') 
])
```

```python
cnn_model.compile(optimizer=Adam(lr=0.01), 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
#Image Augmentation
train_datagen = ImageDataGenerator(
        rescale=1./255, #rescaling pixel value bw 0 and 1
        rotation_range=180,
        vertical_flip=True)
        
        
```

```python
training_set = train_datagen.flow_from_directory(
       './Images/Augmentation_Train/', classes=['Resized_Training_Data'],
        target_size=(360, 480),
        batch_size=32,
        color_mode ="grayscale",
        class_mode='categorical')
```

```python
#Just Feature scaling
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
        './Images/Augmentation_Train/', classes=['Resized_Test_Data'],
        target_size=(360, 480),
        batch_size=32,
        color_mode ="grayscale",
        class_mode='categorical')
```

```python
cnn_model.fit_generator(
        training_set,
        #samples_per_epoch=30, 
        epochs=1,
        validation_data=test_set)
        #validation_steps=379)
```

```python
scores = cnn_model.evaluate_generator(test_set,379) #1514 testing images
print("Accuracy = ", scores[1])
```

```python
history = cnn_model.fit(
    X_train,
    y_train,
    epochs=3,
    shuffle=True,
    validation_split=(0.1),
    verbose=2
)
```

```python
cnn_model.summary()
```

```python
# Evaluate the model using the training data
score = cnn_model.evaluate(X_test, y_test, verbose=2)
print(f"Loss: {score[0]}, Accuracy: {score[1]}")
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='best')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='best')
plt.show()
```
