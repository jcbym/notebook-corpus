<div class="mathbook-content"><h2 class="heading hide-type" alt="Section 16.2 Integral Domains and Fields"><span class="type">Section</span><span class="codenumber">16.2</span><span class="title">Integral Domains and Fields</span></h2><a href="section-domains-and-fields.ipynb" class="permalink">¶</a></div>

<div class="mathbook-content"><p id="p-2335">Let us briefly recall some definitions. If $R$ is a ring and $r$ is a nonzero element in $R\text{,}$ then $r$ is said to  be a <dfn class="terminology">zero divisor</dfn> if there is some nonzero element $s \in R$ such that $rs = 0\text{.}$ A commutative ring with identity is said to be an <dfn class="terminology">integral domain</dfn> if it has no zero divisors.  If an element $a$ in a ring $R$ with identity has a multiplicative inverse, we say that $a$ is a <dfn class="terminology">unit</dfn>. If every nonzero element in a ring $R$ is a unit, then $R$ is called a <dfn class="terminology">division ring</dfn>.  A commutative division ring is called a <dfn class="terminology">field</dfn>.</p></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-gaussian-integers"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.12</span></h6><p id="p-2336">If $i^2 = -1\text{,}$ then the set ${\mathbb Z}[ i ] = \{ m + ni : m, n \in {\mathbb Z} \}$ forms a ring known as the <dfn class="terminology">Gaussian integers</dfn>.  It is easily seen that the Gaussian integers are a subring of the complex numbers since they are closed under addition and multiplication. Let $\alpha = a + bi$ be a unit in ${\mathbb Z}[ i ]\text{.}$ Then $\overline{\alpha} = a - bi$ is also a unit since if $\alpha \beta = 1\text{,}$ then $\overline{\alpha} \overline{\beta} = 1\text{.}$ If $\beta = c + di\text{,}$ then</p><div class="displaymath">
\begin{equation*}
1 = \alpha \beta \overline{\alpha} \overline{\beta} = (a^2 + b^2 )(c^2 + d^2).
\end{equation*}
</div><p>Therefore, $a^2 + b^2$ must either be 1 or $-1\text{;}$ or, equivalently, $a + bi = \pm 1$ or $a+ bi = \pm i\text{.}$  Therefore, units of this ring are $\pm 1$ and $\pm i\text{;}$ hence, the Gaussian integers are not a field. We will leave it as an exercise to prove that the Gaussian integers are an integral domain.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-matrix-z2"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.13</span></h6><p id="p-2337">The set of matrices</p><div class="displaymath">
\begin{equation*}
F =
\left\{
\begin{pmatrix}

0 & 1
\end{pmatrix},
\begin{pmatrix}

1 & 0
\end{pmatrix},
\begin{pmatrix}
0 & 1 \\

\end{pmatrix},
\begin{pmatrix}
0 & 0 \\
0 & 0
\end{pmatrix}
\right\}
\end{equation*}
</div><p>with entries in ${\mathbb Z}_2$ forms a field.</p></article></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-q-sqrt2"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.14</span></h6><p id="p-2338">The set ${\mathbb Q}( \sqrt{2}\, ) = \{ a + b \sqrt{2} : a, b \in {\mathbb Q} \}$ is a field. The inverse of an element $a + b \sqrt{2}$ in ${\mathbb Q}( \sqrt{2}\, )$ is</p><div class="displaymath">
\begin{equation*}
\frac{a}{a^2 - 2 b^2} +\frac{- b}{ a^2 - 2 b^2} \sqrt{2}.
\end{equation*}
</div></article></div>

<div class="mathbook-content"><p id="p-2339">We have the following alternative characterization of integral domains.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="proposition-33"><h6 class="heading"><span class="type">Proposition</span><span class="codenumber">16.15</span><span class="title">Cancellation Law</span></h6><p id="p-2340">Let $D$ be a commutative ring with identity. Then $D$ is an integral domain if and only if for all nonzero elements $a \in D$ with $ab = ac\text{,}$ we have $b=c\text{.}$</p></article><article class="proof" id="proof-103"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2341">Let $D$ be an integral domain. Then $D$ has no zero divisors.  Let $ab = ac$ with $a \neq 0\text{.}$ Then $a(b - c) =0\text{.}$  Hence, $b - c = 0$ and $b = c\text{.}$</p><p id="p-2342">Conversely, let us suppose that cancellation is possible in $D\text{.}$ That is, suppose that $ab = ac$ implies $b=c\text{.}$ Let $ab = 0\text{.}$  If $a \neq 0\text{,}$ then $ab = a 0$ or $b=0\text{.}$  Therefore, $a$ cannot be a zero divisor.</p></article></div>

<div class="mathbook-content"><p id="p-2343">The following surprising theorem is due to Wedderburn.</p></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-77"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">16.16</span></h6><p id="p-2344">Every finite integral domain is a field.</p></article><article class="proof" id="proof-104"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2345">Let $D$ be a finite integral domain and $D^\ast$ be the set of nonzero elements of $D\text{.}$  We must show that every element in $D^*$ has an inverse. For each $a \in D^\ast$ we can define a map $\lambda_a : D^\ast \rightarrow D^\ast$ by $\lambda_a(d) = ad\text{.}$  This map makes sense, because if  $a \neq 0$ and $d \neq 0\text{,}$ then $ad \neq 0\text{.}$  The map $\lambda_a$ is one-to-one, since for $d_1, d_2 \in D^*\text{,}$</p><div class="displaymath">
\begin{equation*}
ad_1 = \lambda_a(d_1) = \lambda_a(d_2) = ad_2
\end{equation*}
</div><p>implies $d_1 = d_2$ by left cancellation. Since $D^\ast$ is a finite set, the map $\lambda_a$ must also be onto; hence, for some $d \in D^\ast\text{,}$ $\lambda_a(d) = ad = 1\text{.}$ Therefore, $a$ has a left inverse. Since $D$ is commutative, $d$ must also be a right inverse for $a\text{.}$ Consequently, $D$ is a field.</p></article></div>

<div class="mathbook-content"><p id="p-2346">For any nonnegative integer $n$ and any element $r$ in a ring $R$ we  write $r + \cdots + r$ ($n$ times) as $nr\text{.}$ We  define the <dfn class="terminology">characteristic</dfn> of a ring $R$ to be the least positive integer $n$ such that $nr=0$ for all $r \in R\text{.}$ If no such integer exists, then the characteristic of $R$ is defined to be 0.  We will denote the characteristic of $R$ by $\chr R\text{.}$ </p></div>

<div class="mathbook-content"><article class="example-like" id="example-rings-zp-field"><h6 class="heading"><span class="type">Example</span><span class="codenumber">16.17</span></h6><p id="p-2347">For every prime $p\text{,}$ ${\mathbb Z}_p$ is a field of characteristic $p\text{.}$ By Proposition <a href="section-mod-n-sym.ipynb#proposition-zn-equiv-classes" class="xref" alt="Proposition 3.4 " title="Proposition 3.4 ">3.4</a>, every nonzero element in ${\mathbb Z}_p$ has an inverse; hence, ${\mathbb Z}_p$ is a field. If $a$ is any nonzero element in the field, then $pa =0\text{,}$ since the order of any nonzero element in the abelian group ${\mathbb Z}_p$ is $p\text{.}$</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="lemma-ring-characteristic"><h6 class="heading"><span class="type">Lemma</span><span class="codenumber">16.18</span></h6><p id="p-2348">Let $R$ be a ring with identity.  If 1 has order $n\text{,}$ then the characteristic of $R$ is $n\text{.}$</p></article><article class="proof" id="proof-105"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2349">If 1 has order $n\text{,}$ then $n$ is the least positive integer such that $n 1 = 0\text{.}$  Thus, for all $r \in R\text{,}$</p><div class="displaymath">
\begin{equation*}
nr = n(1r) = (n 1) r = 0r = 0.
\end{equation*}
</div><p>On the other hand, if no positive $n$ exists such that $n1 = 0\text{,}$ then the characteristic of $R$ is zero.</p></article></div>

<div class="mathbook-content"><article class="theorem-like" id="theorem-integral-domain-characteristic"><h6 class="heading"><span class="type">Theorem</span><span class="codenumber">16.19</span></h6><p id="p-2350">The characteristic of an integral domain is either prime or zero.</p></article><article class="proof" id="proof-106"><h6 class="heading"><span class="type">Proof</span></h6><p id="p-2351">Let $D$ be an integral domain and suppose that the characteristic of $D$ is $n$ with $n \neq 0\text{.}$ If $n$ is not prime, then $n = ab\text{,}$ where $1 \lt a \lt n$ and $1 \lt b \lt n\text{.}$ By Lemma <a href="section-domains-and-fields.ipynb#lemma-ring-characteristic" class="xref" alt="Lemma 16.18 " title="Lemma 16.18 ">16.18</a>, we need only consider the case $n 1 = 0\text{.}$  Since $0 = n 1 = (ab)1 = (a1)(b1)$ and there are no zero divisors in $D\text{,}$ either $a1 =0$ or $b1=0\text{.}$ Hence, the characteristic of $D$ must be less than $n\text{,}$ which is a contradiction.  Therefore, $n$ must be prime.</p></article></div>
