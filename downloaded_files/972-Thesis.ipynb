```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datasets =['Boston Housing','Concrete', 'Energy','Kin8nm']
dataset = datasets[3]

periter_rmse_string = './periter_subset_1adver_rmseResult_' + dataset + '.npy'
periter_ll_string = './periter_subset_1adver_llResult_' + dataset + '.npy'
periter_time_string = './periter_subset_1adver_timeResult_' + dataset + '.npy'

rmse_results = np.load(periter_rmse_string)
ll_results = np.load(periter_ll_string)
time_results = np.load(periter_time_string)
```
