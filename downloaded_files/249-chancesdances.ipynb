```python
#wrap code in a function that can plot any number of logs 
```

```python
def plot_bawps2(wellname):
    
    xaxis = wellname.columns.tolist()
    xaxis = [e for e in xaxis if e not in ('Facies', 'Formation','Well Name', 'Depth')]
    f, ax = plt.subplots(1, len(xaxis), sharey=True, facecolor = 'gray', figsize = (20,10))
    f.suptitle('Bitch Ass Well Plots')
    
    
    for i, x in enumerate(xaxis):
        ax[i].plot(x, 'Depth', data = wellname, color = 'rygbcmk'[i%7])
        ax[i].set(xlabel = x)
        ax[i].grid()
        
            
```

```python
plot_bawps2(shrimplin)
```
