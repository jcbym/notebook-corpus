Simplified static version of Lagerlof's dynamic model. 

#### Model setup

The model is crude but illustrates how changes in relative factor endowments (and/or technology) might have led to broad institutional transitions in response to changing interests of the elite.  An elite (acting as a syndicate) is able to impose whatever institutional arrangement yields it the highest group income.  This elite essentially chooses between these three types of arrangements:

**(1) Egalitarian**: Free access to land, the elite earns a constant share of the total product. We model it here as the average product of labor (implying an equal share) but the tradeoffs below 

$$\frac{F(A,L)}{L}$$

**(2) Slavery/serfdom/tributary agriculture**:  The elite establish control over labor appropriating output and paying workers only a subsistence wage $\underline c$ per unit of labor.To maintain control and avoid uprisings the elite employs 'guards' that earn fraction $\gamma$ of subsistence wage for every person enslaved. The elite's net return: 

$$F(A,L) – \underline c (1+γ)\cdot L$$

**(3) Free Labor:** The elite free labor but enclose lands and pay workers a wage equal to their marginal product. The elite earns profits and land rents: 

$$F(A,L) – w \cdot L$$

where the wage above subsistence is determined competitively $w = \max (MPL(A,L),c)$

We can associate the 'egalitarian' regime with pre-agricultural society 

Belo and in the paper wit is assumed that the production technology is described by a constant returns to scale Cobb-Douglas production function $F(A,L) = A^\alpha L^{1-\alpha}$ where $A$ is effective land (an increment in $A$ can come about by either extending land and/or augmenting it's productivity) and $L$ is labor. 

## Example economy

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('seaborn-poster')
```

```python
c = 0.5  # subsistence wage
a = 0.4  # land share
g = 0.1  # guard pay
```

```python
def F(A,L):
    return A**a * L**(1-a)

def APL(A,L):
    return F(A,L)/L

def MPL(A,L):
    return a*F(A,L)/L

def Y(A,L):
    YE = APL(A,L)
    YS = F(A,L) - (1+g)*c*L
    YF = F(A,L) - MPL(A,L)*L
    YM = np.maximum.reduce([YE, YS, YF])
    return YM, YE, YS, YF
```

```python
def elite_plot(A, L):
    YM, YE, YS, YF = Y(15, L)
    plt.plot(L, YE, linestyle ='-.', label = 'Egalitarian')
    plt.plot(L, YS, linestyle ='--', label='Slave')
    plt.plot(L, YF, label='Free')
    plt.plot(L, YM, label='Max', linewidth=5)
    plt.ylabel('Elite income')
    plt.xlabel('L - Labor')
    plt.legend()
    plt.ylim(0,8)
    plt.show()
```

```python
L = np.linspace(0.1,15, num =100)
elite_plot(15, L)
```

The elite's preferred institutional arrangement depends on the extent of labor-scarcity (population density). It prefers open access at low population densitites, but as population density increases it switches to coerced labor. Coercion allows the elite to capture the full product of the laborers' output while paying them only a subsistence wage plus guards' wages. Finally if population density high enough to push the marginal product of labor (the market wage) below $c(1+γ)$ then the elite prefers free labor.

The model is that clearly rests on some very strong assumptions, for example that the elite manages at all times to impose its preferred institutional arrangement. A cursory glance at history suggests this is not the case.  Slave revolts and serf uprisings were not uncommon and, more importantly, the balance of bargaining power between say landlords and peasants often shifted. While there is considerable support for the idea that slavery and serfdom often often arose in conditions of relative labor scarcity (e.g. Domar, 1970) there are many counter-examples. The population collapse brought about the Black Plague in Europe seems to have contributed to the rise of serfdom in Eastern Europe and Russia, but also to the demise of serfdom in Western Europe. This suggests that the transition from one institutional regime to another is neither mechanical nor simple (Brenner, 1976). 

Lagerlof's published model adds a few other complications. In his model the elite can make further choices, for example hiring or enslaving only an optimum number of workers rather than all as examined here.  It also is dynamic and allows population - growth dynamics.
