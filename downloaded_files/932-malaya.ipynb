```python
import json

with open('test-set-cnn.json') as fopen:
    data = json.load(fopen)['X']
```

```python
print(len(data))
    
from tqdm import tqdm

results = []
batch = 4

try:
    for i in tqdm(range(0, (len(data) // batch) * batch, batch)):
        x = data[i: i + batch]
        x = [f'ringkasan: {row}' for row in x]
        r = pred(x).tolist()
        r = [k.decode() for k in r]
        results.extend(r)
except:
    pass
```

```python
with open('output-small.json', 'w') as fopen:
    json.dump(results, fopen)
```
