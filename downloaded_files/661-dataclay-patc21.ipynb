# Matrix multiplication -- run and analysis

We will be running the matrix multiplication on Mare Nostrum. Feel free to look at `matmul_main.py` (the main application) and the `run_pycompss.sh` (the helper script that realizes the Mare Nostrum reservation and starts the application).

To run the application just connect to Mare Nostrum and run:

```bash
$ cd dataclay-patc21/matrices
$ ./run_pycompss.sh
```

This will give a lot of information, the most relevant one is the last line:

```
Submitted batch job <jobid>
```

You can check your queued jobs with the `squeue` command and cancel them if you have messed up with `scancel <jobid>`

The execution logs can be seen in the file `compss-<jobid>.{out,err}`. You can look at them postmortem or tail them by doing `tail -F compss-<jobid>.{out,err}`. Be patient! You may have to wait a couple of minutes for the job to be accepted. And then it will be a couple more minutes for dataClay to be initialized and COMPSs to launch.

A total of 27 tasks have been executed because the default multiplication uses two 3072x3072 matrices, with 1024x1024 blocks which results (3x3 blocks per matrix, 3x3x3 tasks).

## The dependency graph

You can convert the dependency graph to PDF and get it:

```bash
$ cd ~/.COMPSs/<jobid>/monitor
$ dot -Tpdf complete_graph.dot  > ~/compss_graph.pdf
```

This creates a `compss_graph.pdf` file in the MareNostrum node home, which you can retrieve by doing (**from your laptop/desktop machine**):

```bash
$ scp <user>@mn2.bsc.es:compss_graph.pdf ~/
```

Or use WinSCP or whatever tool you are used to use.
