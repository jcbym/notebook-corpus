```python
# Load the dataset from Hugging Face Hub
dataset_name = "bigcode/the-stack"
subset = "data/jupyter-notebook" 
output_dir = "./downloaded_files"  # Directory to save the downloaded files

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Load the dataset
dataset = load_dataset(
        dataset_name,
        data_dir = subset,
        streaming=True,
        split="train",
    )

# Iterate through the dataset and save each file locally
for idx, sample in enumerate(dataset):
    if idx == 1000:
        break
    
    file_name = str(idx) + "-" + sample['max_stars_repo_name'].split("/")[-1] + ".ipynb" #sample['repo_name'] + " " + sample['path'].split("/")[-1]
    file_content = sample['content']  # Get the file content

    # Save the file locally
    output_path = os.path.join(output_dir, file_name)
    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(file_content)

    print(f"Downloaded {idx+1}: {file_name}")

print("Download completed!")
```
